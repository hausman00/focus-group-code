# Text Mining on Focus Groups

In this lab we will apply the text mining techniques learned in class to a real case: the analysis of a focus group conversation.

The information have been anonymized and no product nor company's name will appear in the text. The focus group was about the switching of a prescribed medicine tamsulosin hydrochloride to over the counter OTC. Experts had conversations on this topic and provided their feedback.

Here we will apply text mining technique to start getting an insight on these conversation.

```python
import nltk
from nltk import FreqDist
from nltk.collocations import *
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

## Preprocessing the conversation

Loading the data. The data has been divided into several files each associated to a particular question easked in any of the two weeks of the workshop.

```python
#Read the dataset
path = 'C:\\Users\\ziadm\\Downloads\\Web Data mining\\'
filePrefix = 'Week1_Q'
dataset={}
dataset_raw = {}
allFeatures=set()
tot_articles = 0
articles_count={}
questions=[1,2,3,4]

N={} # Number of articles in each corpus

for question in questions:
    fileName=path+filePrefix+str(question)+".txt"
    print(fileName)
    f=open(fileName,'r',encoding="utf8")
    text = ''
    text_raw = ''    
    lines=f.readlines()
    tot_articles+=len(lines)
    articles_count[str(question)] = len(lines)
    dataset_raw[str(question)] = list(map(lambda line: line.lower(), lines))
    
    for line in lines:
        text+=line.replace('\n',' ').lower()
        text_raw = line.lower()
    f.close
    N[str(question)]=len(lines)
    
    tokens = nltk.word_tokenize(text)
    dataset[str(question)] = nltk.Text(tokens)
    print(tokens)
```

Removing Punctuations & Stopwording

```python
dataset_clean={}

def remove_punctuation(corpus):
    punctuations = ".,\"-\\/#!?$%\^&\*;:{}=\-_'~()"    
    filtered_corpus = [token for token in corpus if (not token in punctuations)]
    return filtered_corpus

def apply_stopwording(corpus, min_len):
    filtered_corpus = [token for token in corpus if (not token in stopwords.words('english') and len(token)>min_len)]
    return filtered_corpus

#Let's remove punctuation characters and apply stopwording
for question in questions:
    print ("Processing "+ str(question)) 
    dataset_clean[str(question)] = apply_stopwording(remove_punctuation(dataset[str(question)]), 3)
    print (dataset_clean[str(question)])
```

Apply Stemming or Lemmatization

```python
dataset_final={}
all_words=set()

def apply_stemming(corpus):
    stemmer = nltk.PorterStemmer()
    normalized_corpus = [stemmer.stem(token) for token in corpus]
    return normalized_corpus

def apply_lemmatization(corpus):
    lemmatizer = nltk.WordNetLemmatizer()
    normalized_corpus = [lemmatizer.lemmatize(token) for token in corpus]
    return normalized_corpus

for question in questions:
    print ('Processing %s' % str(question))
    dataset_final[str(question)] = apply_lemmatization(dataset_clean[str(question)])
    print (dataset_final[str(question)])
    for token in dataset_final[str(question)]:
        all_words.add(token)

```

### Problem with the collocations statement
Unfortunately the results of the collocations() method cannot be saved in a variable for further processing. We can identify the same collocation using the BigramAssocMeasures with a filter so to ignore all bigrams which occur less than three time. 

Notice how the results strongly resemble the results of the collocations() method.

```python
bigrams = nltk.collocations.BigramAssocMeasures()
finder = BigramCollocationFinder.from_words(text1)
finder.apply_freq_filter(2)
collocations = finder.nbest(bigrams.pmi, 20)
print (collocations)
```

## Selecting Important terms used in the collocations
Notice how these terms are already accounted from the collocation, so no need to calculate tf-idf scores!

```python
import math

def df(category,token):
    # df returns the number of documents containing 'token'
    # Need to use raw data so to have access to individual articles
    articles = dataset_raw[category]
    dfreq = 0
    for article in articles:
        if (token in article):
            dfreq+=1
    return dfreq

def tf(document,token):
    #Simple term frequency within the document
    tokens = nltk.word_tokenize(document)
    return FreqDist(nltk.Text(tokens))[token]

def idf(category,token):
    # idf implementation N[category] contains the number of articles in that category
    return math.log(N[category]*1.0/df(category,token))

def tfidf(category,document,token):
    return tf(document,token)*idf(category,token)

#Load all articles in Question 1
question="1"
importantTerms = set()
threshold = 3
docs = dataset_raw[question]
for doc in docs:
    for terms in collocations:
        for token in terms:
            tfIDF = tfidf(question,doc,token.strip())
            if (tfIDF>threshold):
                print ('%s: tf-idf[%s] tf[%s]' % (token,tfIDF,tf(doc,token)))
                importantTerms.add(token)

print (importantTerms)
```
